```python
import json

def get_password_dic():
    with open("password.json", "r") as f:
        password_str = f.read()
        if password_str == "":
            return {}
        else:
            return json.loads(password_str)

def check_name(name):
    password_dict = get_password_dic()
    if name in password_dict.keys():
        return False
    else:
        return True

def add_password(name, account, password):
    if check_name(name):
        password_dic = get_password_dic()
        password_dic[name] = {
                    "account" : account,
                    "password" : password
                }
        with open("password.json", "w") as f:
            f.write(json.dumps(password_dic, indent=4))    
        return True
    else:
        return False


print("歡迎使用密碼管理器!")

while True:
    mode = input("請問要使用甚麼功能呢?(r:查詢, a:新增, q:離開)").lower()
    if mode == 'q':
        print("謝謝使用!")
        break
    elif mode == 'a':
        name = input("請輸入帳號名稱:")
        account = input("請輸入帳號:")
        password = input("請輸入密碼:")
        if add_password(name, account, password):
            print("新增完成!")
        else:
            print("已有此帳號名稱!")
    elif mode == 'r':
        check = input("請輸入要查詢之帳號名稱:")
        if check == "all".lower():
            password_dict = get_password_dic()  
            for keys in password_dict:
                account =  password_dict[keys]["account"]
                password = password_dict[keys]["password"]
                print(f"帳號名稱:{keys}, 帳號:{account}, 密碼:{password}\n")

        elif check_name(check):
            print("查無此帳號名稱!")

        else:
            password_dict = get_password_dic()
            account =  password_dict[check]["account"]
            password = password_dict[check]["password"]
            print(f"帳號:{account}, 密碼{password}")
    else:
        print("無效輸入!")
```
